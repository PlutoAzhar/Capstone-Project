```python
import sqlite3
```

```python
db=sqlite3.connect("C:/Users/Azhar1/Documents/DataTrained/Shoes.db")

# Declare cursor
cursor = db.cursor()
```

```python
# Create Table1
cursor.execute("CREATE TABLE Table1(ShoeName TEXT, Category TEXT, NoOfColors INT, Price INT)")
```

```python
# Inserting data from the file

with open('Data2.csv','r') as file:
    total_records=0
    for row in file:
        cursor.execute("insert into Table1 values(?,?,?,?)",row.split(","))
        db.commit()
        total_records+=1
print(total_records,'Records inserted')
```

# Join Queries

```python
#Question 1: Write a query find the top rated men's shoes along with their sizes from Table1 and Table3

a=cursor.execute('select Table1.ShoeName, Table3.Size from Table1 LEFT JOIN Table3 ON Table3.Size >=4')
a.fetchall()
```

```python
# Question 2: Write a query that calculates average comfort rating for each category from Table1 and Table3

a=cursor.execute('select Table1.ShoeName, Table3.Comfort from Table1 LEFT JOIN Table3 ON Table3.Comfort')
a.fetchall()
```

```python
# Question 3: Write a query identifies product with a durability/quality/performance rating higher than the average from Table1 and Table3

a=cursor.execute('select Table1.ShoeName, Table3.Durablity from Table1 LEFT JOIN Table3 where Table3.Durablity > (select avg(Durablity) from Table3)')
a.fetchall()
```

```python
# Question 4: Write a subquery finds products with comfort ratings above the average comfort raing using Table1 and Table3

a=cursor.execute('select Table1.ShoeName, Table3.Comfort from Table1 LEFT JOIN Table3 where Table3.Comfort > (select avg(Comfort) from Table3)')
a.fetchall()
```

```python
# Question 5: Write a query joins Table1 and Table2 using Style Code/Product Code column, allowing you to retrieve shoe information along with product details.

a=cursor.execute('select Table1.ShoeName, TableTwo.StyleCode from Table1 LEFT JOIN TableTwo ON TableTwo.StyleCode')
a.fetchall()
```

```python
# Question 6: Write  query that identifies products with a star rating above the average star rating for their respective size.

a=cursor.execute('select Table1.ShoeName, Table3.StarRating from Table1 LEFT JOIN Table3 where Table3.StarRating > (select avg(StarRating) from Table3)')
a.fetchall()
```
