# Scraping Adidas Shoes

```python
#Opening Website

driver = webdriver.Chrome()

driver.get("https://www.adidas.co.in/")

driver.maximize_window()
time.sleep(5)


# Searching for shoes

designation=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[2]/div/header/div[2]/div/div[2]/div/input') 
designation.send_keys('shoes')
time.sleep(2)
search=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[2]/div/header/div[2]/div/div[2]/div/div[2]/div/div[1]/div[2]/a[5]/div[1]/span[2]/strong')
search.click()
time.sleep(3)


# Scraping Name from the page:
AdidasShoeName=[]
a=driver.find_elements(By.XPATH,'//p[@class="glass-product-card__title"]')
for b in a:
    AdidasShoeName.append(b.text)
AdidasShoeName=AdidasShoeName[:10]    
print(AdidasShoeName)


# Scraping Category from the page:
AdidasCategory=[]
a=driver.find_elements(By.XPATH,'//p[@class="glass-product-card__category"]')
for b in a:
    AdidasCategory.append(b.text)
AdidasCategory=AdidasCategory[:10]    
print(AdidasCategory)



# Scraping Category from the page:
AdidasColors=[]
a=driver.find_elements(By.XPATH,'//p[@class="glass-product-card__label"]')
for b in a:
    AdidasColors.append(b.text)
AdidasColors=AdidasColors[:10]    
print(AdidasColors)


# Scraping Category from the page:
AdidasPrice=[]
a=driver.find_elements(By.XPATH,'//div[@class="gl-price-item notranslate"]')
for b in a:
    AdidasPrice.append(b.text)
AdidasPrice=AdidasPrice[:10]    
print(AdidasPrice)

# Scraping url from the page:
url=[]
a=driver.find_elements(By.XPATH,'//a[@data-auto-id="glass-hockeycard-link"]')
for b in a:
    url.append(b.get_attribute('href'))
url=url[:10]    
print(url)

Color=[]
Stylecode=[]
Rating=[]
# Scraping Colors, style code and Rating from the URL's:
for link in url:
    driver=webdriver.Chrome()
    driver.get(link)
    time.sleep(3)
    
    try:
        color=driver.find_element(By.XPATH,'//div[@class="color-label___2hXaD"]')
        Color.append(color.text)
    except exceptions.NoSuchElementException:
        print('--')
    
    try:
        search=driver.find_element(By.XPATH,'(//h2[@class="accordion-title___2sTgR"])[3]')
        search.click()
        time.sleep(2)
        style=driver.find_element(By.XPATH,'//div[@class="content-padding___2kVWQ"]//ul[2]/li[3]')
        StyleCode.append(style.text)
    except exceptions.NoSuchElementException:
        print('--')
    
    try:
        rating=driver.find_element(By.XPATH,'//div[@class="ratings-label-container___13pr-"]')
        Rating.append(rating.text)
    except exceptions.NoSuchElementException:
        print('--')


Adidas=pd.DataFrame({'ShoeName':AdidasShoeName,'Category':AdidasCategory,'No.ofColor':AdidasColors,'Price':AdidasPrice,'Colors':Color,'StyleCode':Stylecode,'Ratings':Rating})
Adidas
```

# Scraping Nike Shoes

```python
#Opening Website

driver = webdriver.Chrome()

driver.get("https://www.nike.com/in/w?q=shoes&vst=shoes")

driver.maximize_window()
time.sleep(5)

# Scraping Name from the page:
ShoeName=[]
a=driver.find_elements(By.XPATH,'//div[@class="product-card__title"]')
for b in a:
    ShoeName.append(b.text)
ShoeName=ShoeName[:10]    
print(ShoeName)


# Scraping Category from the page:
Category=[]
a=driver.find_elements(By.XPATH,'//div[@class="product-card__subtitle"]')
for b in a:
    Category.append(b.text)
Category=Category[:10]    
print(Category)



# Scraping Category from the page:
Colors=[]
a=driver.find_elements(By.XPATH,'//div[@class="product-card__product-count"]')
for b in a:
    Colors.append(b.text)
Colors=Colors[:10]    
print(Colors)


# Scraping Category from the page:
Price=[]
a=driver.find_elements(By.XPATH,'//div[@class="product-price in__styling is--current-price css-11s12ax"]')
for b in a:
    Price.append(b.text)
Price=Price[:10]    
print(Price)

# Scraping url from the page:
url=[]
a=driver.find_elements(By.XPATH,'//a[@class="product-card__link-overlay"]')
for b in a:
    url.append(b.get_attribute('href'))
url=url[:24]    
print(url)

# Scraping Colors, style code and Rating from the URL's:

Color=[]
StyleCode=[]
Rating=[]
for link in url:
    driver=webdriver.Chrome()
    driver.get(link)
    time.sleep(2)
    
    try:
        color=driver.find_element(By.XPATH,'//ul[@class="description-preview__features pt8-sm pb6-sm ncss-list-ul"]//li[1]')
        Color.append(color.text)
    except exceptions.NoSuchElementException:
        print('--')
    
    try:
        style=driver.find_element(By.XPATH,'//ul[@class="description-preview__features pt8-sm pb6-sm ncss-list-ul"]//li[2]')
        StyleCode.append(style.text)
    except exceptions.NoSuchElementException:
        print('--')
        
    try:
        search=driver.find_element(By.XPATH,'//summary[@class=" css-rptnlm"]')
        search.click()
        time.sleep(2)
        rating=driver.find_element(By.XPATH,'//p[@class="d-sm-ib pl4-sm"]')
        Rating.append(rating.text)
    except exceptions.NoSuchElementException:
        print('--')
        
    
print(len(ShoeName),len(Category),len(Colors),len(Price),len(Color),len(StyleCode),len(Rating))
Nike=pd.DataFrame({'ShoeName':ShoeName,'Category':Category,'No.ofColor':Colors,'Price':Price,'Colors':Color,'StyleCode':Stylecode,'Ratings':Rating})
Nike
```

# Scraping Columbia Shoes

```python
#Opening Website

driver = webdriver.Chrome()

driver.get("https://columbiasportswear.co.in/footwear-columbia")

driver.maximize_window()
time.sleep(5)

# Scraping Name from the page:
ShoeName=[]
a=driver.find_elements(By.XPATH,'//h4[@class="product-name"]')
for b in a:
    ShoeName.append(b.text)
ShoeName=ShoeName[:10]    
print(ShoeName)


# Scraping Category from the page:
Price=[]
a=driver.find_elements(By.XPATH,'//span[@class="price-new"]')
for b in a:
    Price.append(b.text)
Price=Price[:10]    
print(Price)


# Scraping url from the page:
url=[]
a=driver.find_elements(By.XPATH,'//h4[@class="product-name"]//span//a')
for b in a:
    url.append(b.get_attribute('href'))   
print(url)

# Scraping Colors, style code and Rating from the URL's:

Color=[]
StyleCode=[]
Rating=[]
for link in url:
    driver=webdriver.Chrome()
    driver.get(link)
    time.sleep(2)
    
    try:
        color=driver.find_element(By.XPATH,'//div[@class="grop-product"]//label')
        Color.append(color.text)
    except exceptions.NoSuchElementException:
        print('--')
        
        
    try:
        style=driver.find_element(By.XPATH,'//div[@class="col-sm-5"]//p[1]')
        StyleCode.append(style.text)
    except exceptions.NoSuchElementException:
        print('--')
        
    try:
        rating=driver.find_element(By.XPATH,'//div[@class="actual-star"]')
        Rating.append(rating.text)
    except exceptions.NoSuchElementException:
        print('--')


print(len(ShoeName),len(Price),len(Color),len(StyleCode),len(Rating))
Columbus=pd.DataFrame({'ShoeName':ShoeName,'Price':Price,'Color':Color,'StyleCode':StyleCode,'Ratings':Rating})
Columbus
```
